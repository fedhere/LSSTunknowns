```python
Fields['inGP'] = np.isin(Fields['pixId'].values, ipix_GP)
Fields['inLMC'] = np.isin(Fields['pixId'].values, ipix_LMC)
Fields['inSMC'] = np.isin(Fields['pixId'].values, ipix_SMC)

ms = np.stack( [Fields['inGP'].values,Fields['inLMC'].values, Fields['inSMC'].values], axis=1)
Fields['other'] = np.sum(ms, axis=1)==0
```

get Nfields and Nstars

```python
msname = 'GP_allsky'
dataRaw = df_GP_allsky
dicNfields = {}
dicNstar = {}

dicNfields['db'] = dblist_all
dicNstar['db'] = dblist_all

for fltpair in fltpairs:  
    dicFoMtGaps[fltpair] = []
    dicNfields[fltpair] = []
    dicNstar[fltpair] = []
    
    for dbname in dblist_all: 

        df = dataRaw[fltpair][dbname]
        #df['starDensity'] = starDens[ df['pixId'].values ] 
        df['starDensity'] = starDens[ df['pixId'].values ] 
        idx = df['Nv'] > Nvth[fltpair]
        Nfields = df['Nv'][idx].shape[0]
        Nstar = np.sum( df['starDensity'][idx] )
                
        dicNfields[fltpair].append(Nfields)
        dicNstar[fltpair].append(Nstar)

        print('>>>', fltpair, dbname, Nfields, Nstar)
    
    df_fom = pd.DataFrame(dicNfields)
    df_fom.to_csv('./data/df_Nfields_{}.csv'.format(msname), index=False)
    
    df_fom = pd.DataFrame(dicNstar)
    df_fom.to_csv('./data/df_Nstars_{}.csv'.format(msname), index=False)
```
